NumPy data types (dtype) define how data is stored in memory. Choosing the right dtype improves performance, memory usage, and model speed.

- NumPy dtypes are fixed-size

- Smaller dtype → faster & memory-efficient

- Wrong dtype → slow ML pipeline

- float32 is king in AIML

| Use Case    | dtype        |
| ----------- | ------------ |
| Features    | `float32`    |
| Labels      | `int32`      |
| Images      | `uint8`      |
| Masks       | `bool`       |
| Time series | `datetime64` |
