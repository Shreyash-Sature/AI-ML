```python
import numpy as np
```

```python
#Creatig arrays using built-in-functions
#arrays with zeros
arr1 = np.zeros((2,3)) #prefill
print(arr1,arr1.shape)
```

```python
#array with ones
arr2 = np.ones((2,3), dtype = "int64")
print(arr2,arr2.shape)
```

# Creatiing array with prefilling number 

```python
"""
Synatx : 
arr_name = np.full((n,m), num_to_fill)
n,m : are dimensions
"""
arr3 = np.full((3,4), 100) #prefill with value
print(arr3,arr3.shape)
```

# Creating an identity matrix
Identity matrix is a square matrix with diagonals elements "1".

```python
arr4 = np.eye(3)
print(arr4, arr4.shape)
```

```python
arr5 = np.arange(0,12,2)
print(arr5, arr5.shape)
```

```python
#creating evenly spaced array
arr6 = np.linspace(1,100,4)
print(arr6, arr6.shape)
```
