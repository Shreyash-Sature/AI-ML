# 3D Arrays
- (axis 0, axis 1, axis 2)
- = (depth, rows, columns)


```python
import numpy as np
```

```python
#lets create a 3D array

"""
steps : to create a (2 x 2 x 3) array

arr3D = np.array()
arr3D = np.array([])
arr3D = np.array([ [],[] ])
arr3D = np.array([[[a,c,s],[q,w,e]],[[a,c,s],[q,w,e]]])

"""

arr3D = np.array([[[1,2,3],
                   [4,5,6]],
                  
                  [[4,6,8],
                   [0,3,8]]]) 
print(arr3D)
```

# Indexing on 3D array

```python
# indexing on 3D arrays
#lets search a key = 4 from array 
key = arr3D[0,1,0]
print(key)

#lets search a key = 8 from array 
key = arr3D[1,1,2]
print(key)
```

# Slicing on 3D array
- General Slicing Syntax
- array[ depth , rows , columns ]


```python
# slice 1st row from both layers
print(arr3D[:, 0, :])
```

```python
# now slice 1st colum
print(arr3D[:,:,0])
```

```python
print(arr3D[0, 0:1, 0:2])
```

# manupulating data from 3D arrays

```python
arr3D[:,0,:] = 1 #replaing all values of 1st rows with 1
print(arr3D)
```
