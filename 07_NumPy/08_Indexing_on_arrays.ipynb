# Indexing on arrays (1D and 2D arrays)

```python
import numpy as np
```

```python
#Lets create a array
arr = np.array([1,2,5,8])
print(arr)
```

```python
# Indexing 
#accessing elements from array using index
print(arr[1]) 
```

```python
# Indexing on 2D array
arr2 = np.array([[1,3,2],[3,6,9],[2,1,6]])
print(arr2)

#lets access element 2 from 2D array
print(arr2[2][0])
```

```python
# Fancy Indexing

#accessing multiple elements at a same time from a 1D array

arr3 = np.array([1,2,5,8,0,3,6])

indx = [0,3,5]
print(arr3[indx])
```

```python
#Boolean Indexing
#pass the condition at place of index
print(" Elements greater than 3 from array are : ",arr3[arr3 > 3]) 
print(" Even Numbers from the array : ",arr3[arr3 % 2 == 0])#even numbers
```
