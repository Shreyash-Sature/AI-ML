# Vectorization
- Vectorization means performing operations on entire arrays at once, without explicit Python loops.
- uses low-level C optimizations inside NumPy.

```python
import numpy as np
```

```python
arr = np.array([1,2,3])
arr2 = np.array([3,5,7])

print(arr ** 2)
print(arr + arr2)
```
