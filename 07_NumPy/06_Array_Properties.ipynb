```python
import numpy as np
```

```python
#lets create a 2d array
arr = np.array([[1,2,3],[2,5,2],[5,0,2],[1,2,3]])
print(arr)
```

```python
# dimensions of array 
#(arr.shape)
print(arr.shape)
```

```python
# size of array (no. of elements)
print(arr.size)
```

```python
#dtype of array
print(arr.dtype)
```

```python
# total number of dimensions
print(arr.ndim)
```

# Converting existing array by type converting into a new array
eg.: int -> float

```python
arr_float = arr.astype(np.float64)
print(arr_float, arr_float.dtype)
```
