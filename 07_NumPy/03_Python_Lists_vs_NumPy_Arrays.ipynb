## Python Lists vs NumPy arrays

1) NumPy arrays are faster (Optimized C)
2) NumPy stores data in contegius memory blocks
3) Uses less memory(optimized storage)
4) Built in mathematical functions
5) Supports direct victorized operations

```python
import numpy as np
import time
```

```python
# Comparosion in terms of execution performance

size = 100_000_000

#python_list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"Time taken by python list is {end - start} seconds")

#numpy arrays

np_arr = np.array(py_list)
start = time.time()
#vectorization
sq_arr = np_arr ** 2
end = time.time()
print(f"Time taken by numpy array is {end - start} seconds")
```
