## Vectorization

- Vectorization means applying operations on entire NumPy arrays at once without using loops.
  
- Vectorization is a technique where operations are performed on entire arrays at once instead of using explicit Python loops.

- In NumPy, vectorization allows computations to be executed at the C level, which makes programs faster and more efficient.
