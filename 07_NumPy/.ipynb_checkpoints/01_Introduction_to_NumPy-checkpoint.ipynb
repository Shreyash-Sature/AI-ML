## NumPy (Numerical Python)

NumPy is a Python library used for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays efficiently.

The main object in NumPy is the ndarray (N-dimensional array).

- Faster than Python lists (written in C)
- Uses less memory
- Essential for AI, ML, Data Science, Image Processing
- Base library for Pandas, OpenCV, SciPy, TensorFlow

| Feature    | Python List | NumPy Array |
| ---------- | ----------- | ----------- |
| Speed      | Slower      | Faster      |
| Memory     | More        | Less        |
| Operations | Limited     | Vectorized  |
| Dimensions | 1D mostly   | Multi-D     |
