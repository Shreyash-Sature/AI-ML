# Operations on array
1) Reshaping Array :
   - changing the shape (dimensions) of an array without changing its data.
   - eg .:
       1) (3,2) -> (2,3)
       2)   1D -> 2D
       3) 2D -> 1D

```python
import numpy as np
```

```python
#lets create a 2D array with dimensions (3,2)
arr = np.array([[1,2,3],[3,4,6]])
print(arr)
print(f"Dimensions of numpy array are : {arr.shape}")
```

```python
#Now reshape array 
arr1 = arr.reshape(3,2)
print(arr1)
print(f"Dimensions of numpy array are : {arr1.shape}")
```

Flatten array 
- 2D -> 1D

```python
arr2 = arr1.flatten() 
print(arr2)
print(f"Dimensions of numpy array are : {arr2.shape}")
```
