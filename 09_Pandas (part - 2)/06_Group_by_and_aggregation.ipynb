```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data .csv")
df2 = df.copy()
```

```python
#groups the same type of data and we can calculate or perform operation on that grouped data
#Syntax : 
# df.groupby("parameter_on_baisis we group data")
df2.groupby("country")["income"].mean()
df2.groupby("country")["income"].min()
```

```python
#aggeragte
df2.groupby("country")["income"].agg(["min", "max","mean"])
#OR
df2.groupby("country")["income"].agg(min_salary="min", max_salary = "max", mean_salary = "mean")
```

```python
df2.groupby("country").agg(
    max_salary = ("income","max"),
    avg_age = ("age", "mean")
)
```
