```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data .csv")
df2 = df.copy()
```

```python
#renaming all coulumn names

df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income"]
df2
```

```python
# renaming a single column name or row name
df2.rename(columns = {"Income" : "Salary"})
df2.rename(index = {1 : "First"})
```

```python
# Function for Sorting values
sorted_df = df2.sort_values(["Income","Age"])

sorted_df
```

```python
sorted_df.sort_index()
```

```python
#Ranking
sorted_df["Ranking"] = sorted_df["Income"].rank()
sorted_df
```
