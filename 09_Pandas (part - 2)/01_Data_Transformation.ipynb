# Transforming Data (Featuring data):
- .apply(fnc)
- .map()
- replace(old,new)


```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data .csv")

df2 = df.copy()

# now create a new column of tax

df2["tax"] = df2["income"].apply(lambda x : "20%" if x >=50000 else "10%" )
```

```python
df2
```

```python
# .map() : map existing values with diff values
# change gender full form to short form

gender_map = {
    "Male" : "M",
    "Female" : "F",
    "Unknown" : "U"
}

df2["gender"] = df2["gender"].map(gender_map)
```

```python
df2
```

```python
df2.assign(new_income = df2["income"] * 1.1)
```

```python
# replace(old,new)
#replaces values in specific values from specific column 
df2["country"] = df2["country"].replace("USA","US")
```
