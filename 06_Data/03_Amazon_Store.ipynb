## Amazon Store : Clean and Structure Data

- data is inconsistant : mixed types(rating)
- missing data (age)
- Duplicate data 

```python
import json
```

```python
# Load the data 
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("02_Store_Data.json")
print(data)
print(type(data))
```

```python
# Cleaning and Structuring Data

def clean_data(data):
    cleaned_data = []
    unique_user = set()
    text_to_num = {"one" : 1,
                   "two" : 2,
                   "three" : 3,
                   "four" : 4,
                   "five" : 5
                  }
    for user in data:
        #clean ratings
        raw_rating = str(user["rating"]).strip().lower()
       

        if (raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating


        #Handling null values
        raw_age = user.get("age")  #.get() : gives value if present and doesnt throw error if value isn't present
        if(raw_age == None):
            user["age"] = None

        #Deduplication 

        if(user["name"] in unique_user):
            continue
        unique_user.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
        
```

```python
data = clean_data(data)
```

## Getting meaningful insights from data
1) Whats average of rating?
2) Whats % of poor (rating < 3) rating?

```python
# Getting meaningful insights from data

def get_insights(data):

    #avg rating
    counter = 0
    total_rating = 0
    for user in data :
        counter += 1
        total_rating += float(user["rating"])
        
    avg_rating = total_rating / counter

    return avg_rating
```

```python
avg = get_insights(data)
print(avg)
```

```python
#Recomendation 
def get_recommendation(data):
    recommendations = []

    for user in data:
        curr_recom = {}
        curr_recom["name"] = user["name"]
        
        if(float(user["rating"])) > 4:
            curr_recom["brand"] = "Apple"
        else :
            curr_recom["brand"] = "Samsung"

        recommendations.append(curr_recom)
    return recommendations
            
```

```python
get_recomendation(data)
```
