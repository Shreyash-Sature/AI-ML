# Series Properties :
1) Homogenous (same datatype)
2) Vectorized Operations
3) Handling missing values with NaN
4) Mutable Values, immutable size

```python
import pandas as pd
```

```python
# 1) Homogenous
s = s = pd.Series([11,12,13,14,'Abc'], index =['Shreyash','Kalu','Dholu','Bholu',25])
print(s)
print(type(s))
```

```python
#2) Vectorized Operations

s1 = pd.Series([1,2,3,4])
s2 = pd.Series([10,11,12,13])

print(s1 + s2)
```
