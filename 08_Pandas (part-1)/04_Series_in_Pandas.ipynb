# Series 
- A Pandas Series is a one-dimensional labeled array capable of holding data of any type (integers, floats, strings, objects, etc.).

  ðŸ”¹ Key Characteristics of a Series

1) One-dimensional data structure

2) Has values and an associated index

3) Index can be integer-based or custom labels

4) Mutable (values can be changed)

```python
import pandas as pd
```

```python
s = pd.Series([11,23,43,54,56])
print(s)
print(type(s))

#index

print(s[0])
```

- here 0,1,2,3,4 are labels
- labels are named index

```python
s = pd.Series([11,12,13,14], index =['Shreyash','Kalu','Dholu','Bholu'])
print(s)

print(s['Kalu']) #can be accessed by name of label
```

```python
print(s[1])# can also be accessed by index
```
