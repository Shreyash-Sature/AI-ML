# Data science process
1) Define Problem / Understand problem
2) Data Collection
3) EDA
4) Data Cleaning (preprocessing)
5) Featuring Data
6) Model Building
7) Model Evaluation
8) Model eployment

# 1. Problem Understanding (Business Understanding)

- Goal: Clearly understand what problem you are solving.

Key points:

- Define the objective (prediction, classification, recommendation, insight)

- Understand business/domain context

- Decide success metrics (accuracy, RMSE, profit, etc.)

Example:

Predict whether a customer will churn in the next 30 days.

# 2. Data Collection

- Goal: Gather relevant data from various sources.

Common data sources:

- Databases (SQL, NoSQL)

- CSV / Excel files

- APIs

- Web scraping

- Sensors / Logs

Example:

Customer details

Transaction history

Website usage logs

# 3. Data Understanding / Exploration (EDA)

- Goal: Understand the data before cleaning or modeling.

Activities:

- Check data size & structure

- Identify missing values

- Detect outliers

- Study distributions and correlations

Tools:

- Pandas, NumPy

- Matplotlib, Seaborn

Output:

Insights like trends, patterns, anomalies

# 4️. Data Cleaning (Data Preprocessing)

- Goal: Make data usable and reliable.

- Common tasks:

Handle missing values (mean, median, drop)

Remove duplicates

Fix inconsistent data

Handle outliers

Correct data types

- Example:

Replace missing age values

Remove invalid records

# 5️. Feature Engineering

- Goal: Create meaningful input features for models.

- Techniques:

Feature selection

Feature transformation

Encoding categorical variables

Scaling / normalization

Creating new features

- Example:

Convert date → day/month/year

Create “total purchase value”

# 6️. Model Building

- Goal: Train machine learning models on prepared data.

- Common models:

Linear Regression

Logistic Regression

Decision Trees

Random Forest

SVM

Neural Networks

- Steps:

Split data (Train/Test)

Train model

Tune hyperparameters

# 7️. Model Evaluation

- Goal: Check how well the model performs.

- Metrics:

Classification: Accuracy, Precision, Recall, F1-score

Regression: RMSE, MAE, R²

Clustering: Silhouette score

- Example:

Accuracy = 92%, Recall = 88%

# 8️. Model Deployment

- Goal: Make the model usable in real-world applications.

Deployment methods:

Web apps (Flask, FastAPI)

Cloud platforms

APIs

Embedded systems

- Example:

Churn prediction API for customer support team
