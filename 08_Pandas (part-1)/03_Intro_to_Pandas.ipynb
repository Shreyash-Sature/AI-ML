# Pandas : 
- Pandas is a powerful Python library used for data manipulation, data analysis, and data cleaning.
- It is built on top of NumPy and provides high-level, easy-to-use data structures that make working with structured data fast and intuitive.

Pandas provides:

- Fast and flexible data structures

- Tools to read, clean, transform, analyze, and visualize data

- Support for tabular data (like Excel, CSV, SQL tables)

# Core Data Structures in Pandas
# 1) Series

- One-dimensional labeled array

- Similar to a column in Excel

```python
import pandas as pd
seri = pd.Series([10,20,30])
print(seri)
```

# 2) DataFrame

- Two-dimensional data structure (labeled array) (rows + columns)

- Similar to an Excel sheet or SQL table

```python
#let we have some data in form of dictionary
info = {
    "Names" : ["Amit", "Smruti", "Kalu"],
    "CGPA" : [8.9, 9.3, 9.5],
    "Marks" : [90, 93, 97]
    
}
df = pd.DataFrame(info)
print(df)
```
