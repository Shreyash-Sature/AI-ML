# Interview questions

1. What is EDA?

EDA is the process of analyzing and visualizing data to understand patterns, detect anomalies, and prepare data for modeling.

2. Why is EDA important before modeling?

It helps identify missing values, outliers, relationships, and incorrect assumptions which can negatively affect model performance.

3. Difference between EDA and Data Cleaning?
| EDA                   | Data Cleaning          |
| --------------------- | ---------------------- |
| Identify problems     | Fix problems           |
| Visualization focused | Transformation focused |
| Insight generation    | Data correction        |


4. What plots are used in EDA?

Histogram

Boxplot

Scatter plot

Heatmap

Pairplot

5. How do you handle missing values?

Remove rows/columns

Mean/median/mode imputation

Predict missing values

6. What are outliers and how do you detect them?

Outliers are extreme values detected using boxplots, IQR, or Z-score methods.

7. What is multicollinearity?

When independent variables are highly correlated, affecting model stability.

8. Can EDA improve model accuracy?

Yes, by selecting better features, removing noise, and correcting data issues.
