# DataFrames in pandas
- A DataFrame is a 2-dimensional, labeled, size-mutable data structure that can store different types of data (int, float, string, boolean).

```python
import pandas as pd
```

```python
#dataFrame

info = {
    "Name" : ["Shreyash","Samir","Soya"],
    "Age" : [19, 20, 21],
    "CGPA" : [8.7, 8.9, 7.1]
}

df = pd.DataFrame(info)
print(df)
print(type(df))

print(df.index)
print(df.columns)
```

```python
df
```

```python
#creating DataFrame from List of List
df1 = pd.DataFrame([["Shrey",20],["Sumit",21],["Kalu",22]], columns =["Name","Age"])
print(df1)
```

```python
df1
```

```python
#Creating DataFrame from NumPy array

import numpy as np

arr = np.array([[1,2,3],[4,5,6]])

df = pd.DataFrame(arr, columns = ["A","B","C"])

print(df)
```
