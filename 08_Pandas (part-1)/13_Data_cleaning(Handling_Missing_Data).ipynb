# Data cleaning (Handling missing data)

Functions used to handle missing values
- isnull() / isna()
- isnull().sum()
- dropna()
- fillna(value)
- ffill()
- bfill()

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data .csv")
df.isnull()
```

```python
df.isnull().sum() #counts the number of missing values per column
```

```python
df.dropna()
df.dropna(axis=1)
```

# replacing missing values 


```python
# synatx : df.["col_name"].fillna(Value_to_fill)
```

```python
cleaned_data = df.copy() #copy

avg_age = cleaned_data["age"].mean() #avg_age

cleaned_data["age"].fillna(avg_age) 
```

```python
# forward fill (ffill())
#fills value from last

df.ffill()
```

```python
df.bfill() #backward fill
```
