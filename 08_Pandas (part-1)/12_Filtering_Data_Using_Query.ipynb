```python
import pandas as pd
```

# Always make copy and not view 
- use : .copy()

```python
df = pd.read_csv("globalAirQuality.csv")
```

# query() Rules 
- Expression as string
- column based reference
- Backticksor col_name with space or spacial characters
- @ to refernce python vars
- Chained comparisions

```python
df.query("aqi > 200") 
```

```python
df.query("aqi > 200 & temperature > 10")[["city","aqi"]]
```

```python
#python variables
aqi_value = 200
aqi_data = df.query("aqi > @aqi_value & temperature >30")[["city","aqi"]].copy() 
print(aqi_data)
```
