# Handling duplicte values
- duplicated() : used to detect duplicate
- drop_duplicates() : used to drop duplicate

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data .csv")
```

```python
df.duplicated()
```

```python
df["country"].duplicated() #to check if  country is duplicated
```

```python
df[["country","gender"]].duplicated() #to check if combo of country and gender is duplicated
```

```python
df2 = df.copy()
df.drop_duplicates(inplace = True) #makes changes in original dataFrame
```
